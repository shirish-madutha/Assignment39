```python
""" Q1. Load the flight price dataset and examine its dimensions. How many rows and columns does the
dataset have? """

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_excel("Data_Train.xlsx")
shape=df.shape
print("Dimensions :",shape)
print("Number of rows are :",shape[0])
print("Number of columns are :",shape[1])
```

```python
""" Q2. What is the distribution of flight prices in the dataset? Create a histogram to visualize the
distribution. """

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x="Price", kde=True)

# obervations
# The distribution of flight prices in the dataset is right skewed distribution.
```

```python
""" Q3. What is the range of prices in the dataset? What is the minimum and maximum price? """

price_range = df["Price"].max() - df["Price"].min()
print("Price Range:", price_range)
print("Minimum price is",df["Price"].min())
print("Maximum price is",df["Price"].max())
```

```python
""" Q4. How does the price of flights vary by airline? Create a boxplot to compare the prices of different
airlines. """

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="Price", y="Airline")
```

```python
""" Q5. Are there any outliers in the dataset? Identify any potential outliers using a boxplot and describe 
how they may impact your analysis. """

plt.figure(figsize=(10,6))
sns.boxplot(data=df, x="Price")

# ans
""" Impact on analysis:
Outliers can significantly impact your analysis in various ways:

Distorting statistical measures: Outliers can skew the mean, making it no longer representative of the central 
tendency of the data. The median is less affected by outliers compared to the mean.

Influence on correlations: Outliers can create or disrupt relationships between variables, leading to misleading 
correlation coefficients.

Regression analysis: Outliers can significantly affect the coefficients and interpretation of regression models,
leading to less accurate predictions.

Assumption violations: Many statistical tests assume that the data is normally distributed and free of outliers.
Violating these assumptions can lead to incorrect conclusions.

Data integrity: Outliers may indicate data entry errors or measurement errors. It's essential to verify the 
accuracy of these data points before including or excluding them from the analysis.

Visualization: Outliers can distort visualizations, making it challenging to interpret and understand the 
underlying patterns in the data. """
```

```python
""" Q9. Load the Google Playstore dataset and examine its dimensions. How many rows and columns does
the dataset have? """

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv("playstore_dataset.csv")
shape=df.shape
print("Dimensions :",shape)
print("Number of rows are :",shape[0])
print("Number of columns are :",shape[1])
```

```python
""" Q10. How does the rating of apps vary by category? Create a boxplot to compare the ratings of different
app categories. """

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 8))  

top_10_categories = df['Category'].value_counts().nlargest(10).index
sns.boxplot(x='Category', y='Rating', data=df[df['Category'].isin(top_10_categories)], palette='Set3')

plt.xticks(rotation=45)
plt.xlabel('App Category')
plt.ylabel('Ratings')
plt.title('Comparison of App Ratings by Category')

plt.show()
```

```python
""" Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps
with the highest number of installs. """

# Create a frequency table for the number of installs for each app
app_installs_freq = df.groupby('App')['Installs'].sum().reset_index()

# Sort the frequency table in descending order to get the top 10 most popular apps
top_10_apps = app_installs_freq.nlargest(10, 'Installs')

print(top_10_apps)

```
